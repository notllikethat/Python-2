## 1. Most popular name in each state
* Make visualization explaining which name was the most popular in each state before and after 1990 year

**Before 1990 year:**

```python
before_1990_year = StateNames[StateNames.Year <= 1990]
before_1990_year = before_1990_year[["Name", "State", "Count"]]
before_1990_year = before_1990_year.groupby(["Name", "State"], as_index = False).aggregate({"Count" : sum})
before_1990_year = before_1990_year.loc[before_1990_year.groupby("State")["Count"].idxmax()]

before_1990_year['Name'] = before_1990_year[['State', 'Name']].agg('-'.join, axis = 1)
before_1990_year = before_1990_year[['Name', 'Count']]

before_1990_year
```

```python
before_1990_year_plot = before_1990_year.plot.bar(x = 'Name', y = "Count", rot = 90, 
                                                  figsize = (15, 5), color = 'cornflowerblue')
before_1990_year_plot.set_title('Popular names before 1990 year')
before_1990_year_plot.set_ylabel('Count')
before_1990_year_plot.set_xlabel('Names')

before_1990_year_plot
```

**after 1990 year:**

```python
after_1990_year = StateNames[StateNames.Year > 1990]
after_1990_year = after_1990_year[["Name", "State", "Count"]]
after_1990_year = after_1990_year.groupby(["Name", "State"], as_index = False).aggregate({"Count" : sum})
after_1990_year = after_1990_year.loc[after_1990_year.groupby("State")["Count"].idxmax()]

after_1990_year['Name'] = after_1990_year[['State', 'Name']].agg('-'.join, axis = 1)
after_1990_year = after_1990_year[['Name', 'Count']]

after_1990_year
```

```python
after_1990_year_plot = after_1990_year.plot.bar(x = 'Name', y = "Count", 
                                                rot = 90, figsize = (15, 5), color = 'cornflowerblue')
after_1990_year_plot.set_title('Popular names after 1990 year')
after_1990_year_plot.set_ylabel('Count')
after_1990_year_plot.set_xlabel('Names')

after_1990_year_plot
```

## 2. Top-3 most popular names in 1990-2000 years in three states
* Make visualization explaining which are top 3 of the most popular names in 1990-2000 years separately among boys and girls in three states selected by you

I selected 3 following states of USA: 
* State of New York, 
* State of Texas,
* State of California.

### State of New York

```python
years_1990_2000 = StateNames[(StateNames.Year >= 1990) & (StateNames.Year <= 2000)]
New_York = years_1990_2000[years_1990_2000.State == 'NY']

New_York_female = New_York[New_York.Gender == 'F']
New_York_female = New_York_female.groupby(['Name', 'Gender']).aggregate({'Count' : 'sum'})
New_York_female = New_York_female.sort_values('Count', ascending = False)[:3]

New_York_male = New_York[New_York.Gender == 'M']
New_York_male = New_York_male.groupby(['Name', 'Gender']).aggregate({'Count' : 'sum'})
New_York_male = New_York_male.sort_values('Count', ascending = False)[:3]

New_York_popular_names = pd.concat([New_York_male, New_York_female])
New_York_popular_names
```

```python
New_York_popular_names_plot = New_York_popular_names.plot.bar(color = 'cornflowerblue', figsize = (10, 5))
New_York_popular_names_plot.set_title('Three most popular names in New York between 1990 and 2000')
New_York_popular_names_plot.set_ylabel('Count')
New_York_popular_names_plot.set_xlabel('Name and gender')

New_York_popular_names_plot
```

### State of Texas

```python
years_1990_2000 = StateNames[(StateNames.Year >= 1990) & (StateNames.Year <= 2000)]
Texas = years_1990_2000[years_1990_2000.State == 'TX']

Texas_female = Texas[Texas.Gender == 'F']
Texas_female = Texas_female.groupby(['Name', 'Gender']).aggregate({'Count' : 'sum'})
Texas_female = Texas_female.sort_values('Count', ascending = False)[:3]

Texas_male = Texas[Texas.Gender == 'M']
Texas_male = Texas_male.groupby(['Name', 'Gender']).aggregate({'Count' : 'sum'})
Texas_male = Texas_male.sort_values('Count', ascending = False)[:3]

Texas_popular_names = pd.concat([Texas_male, Texas_female]).sort_values('Count', ascending = False)
Texas_popular_names
```

```python
Texas_popular_names_plot = Texas_popular_names.plot.bar(color = 'cornflowerblue', figsize = (10, 5))
Texas_popular_names_plot.set_title('Three most popular names in Texas between 1990 and 2000')
Texas_popular_names_plot.set_ylabel('Count')
Texas_popular_names_plot.set_xlabel('Name and gender')

Texas_popular_names_plot
```

### State of California

```python
years_1990_2000 = StateNames[(StateNames.Year >= 1990) & (StateNames.Year <= 2000)]
California = years_1990_2000[years_1990_2000.State == 'CA']

California_female = California[California.Gender == 'F']
California_female = California_female.groupby(['Name', 'Gender']).aggregate({'Count' : 'sum'})
California_female = California_female.sort_values('Count', ascending = False)[:3]

California_male = California[California.Gender == 'M']
California_male = California_male.groupby(['Name', 'Gender']).aggregate({'Count' : 'sum'})
California_male = California_male.sort_values('Count', ascending = False)[:3]

California_popular_names = pd.concat([California_male, California_female]).sort_values('Count', ascending = False)
California_popular_names
```

```python
California_popular_names_plot = California_popular_names.plot.bar(color = 'cornflowerblue', figsize = (10, 5))
California_popular_names_plot.set_title('Three most popular names in California between 1990 and 2000')
California_popular_names_plot.set_ylabel('Count')
California_popular_names_plot.set_xlabel('Name and gender')

California_popular_names_plot
```
